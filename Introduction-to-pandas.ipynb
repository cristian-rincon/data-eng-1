## Data wrangling

```python
# 1. Add newspaper_iud to DataFrame

csv_test_et['newspaper_iud'] = 'eltiempo'
csv_test_et
```

```python
# 2. Get host
from urllib.parse import urlparse

csv_test_et['host'] = csv_test_et['url'].apply(lambda url: urlparse(url).netloc)
csv_test_et
```

```python
csv_test_et['host'].value_counts()
```

# Missing Data

```python
# 3. Fill missing data

missing_titles_mask = csv_test_et['title'].isna()

# missing_titles_mask

missing_titles = (csv_test_et[missing_titles_mask]['url']
                     .str.extract(r'(?P<missing_titles>[^/]+)$')
                     .applymap(lambda title: title.split('-'))
                     .applymap(lambda title_word_list: ' '.join(title_word_list)))
missing_titles
```

# Additional cleanup

```python
# 4. Add uid to files
import hashlib

uids = (csv_test_et
        .apply(lambda row: hashlib.md5(bytes(row['url'].encode())), axis=1)
        .apply(lambda hash_object: hash_object.hexdigest())
       )

csv_test_et['uid'] = uids
csv_test_et.set_index('uid', inplace=True)

csv_test_et
```

```python
stripped_body = (csv_test_et
                .apply(lambda row: row['body'],axis=1)
                .apply(lambda body: list(body))
                .apply(lambda letters: list(map(lambda letter: letter.replace('\n', ''), letters)))
                .apply(lambda letters: ''.join(letters))
                )

stripped_body
```

# Data enrichment

```python
# 6. Tokenize title and body
import nltk
from nltk.corpus import stopwords

# conda install nltk
# run this lines first time you run this notebook on your environment
# nltk.download('punkt')
# nltk.download('stopwords')

stop_words = set(stopwords.words('spanish'))

def tokenize_column(df, column_name):
    return (df
               .dropna()
               .apply(lambda row: nltk.word_tokenize(row[column_name]), axis=1)
               .apply(lambda tokens: list(filter(lambda token: token.isalpha(),tokens)))
               .apply(lambda tokens: list(map(lambda token: token.lower(), tokens)))
               .apply(lambda word_list: list(filter(lambda word: word not in stop_words, word_list)))
               .apply(lambda valid_word_list:  len(valid_word_list))
           )
csv_test_et['n_tokens_title'] = tokenize_column(csv_test_et,'body')

csv_test_et
```
