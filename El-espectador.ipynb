```python
import requests

el_espectador = 'https://elespectador.com'

response = requests.get(el_espectador)
```

```python
print(response.text)
```

## Extracting information from newspaper page

```python
import bs4

soup = bs4.BeautifulSoup(response.text, 'html.parser')
# print(soup.title.text)
# print(soup.prettify())
soup.h2.a
```

```python
# soup.find_all('h2')
```

```python
news_links = soup.select('h1 > a , h2 > a')
# print(news_links)
news = [item['href'] for item in news_links]

for item in news:
    print(item)
```
